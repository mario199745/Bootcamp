```python
## Construcción de modelo NLP con Tensorflow y Keras
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
# Cargar un conjunto de datos de críticas de películas
# (asegúrate de tener un conjunto de datos etiquetado)
# En este ejemplo, utilizamos un conjunto de datos ficticio.
reviews = ["Esta película es genial.", "No me gustó la trama.", "Actuaciones impresionantes.", "Aburrida, no la recomendaría."]
labels = [1, 0, 1, 0]  # Etiquetas de sentimiento (1: positivo, 0: negativo)
```

```python
# Tokenización y secuenciación de texto
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")

# Tokenizer: La clase Tokenizer de Keras se utiliza para convertir texto en secuencias de números enteros. 
# Se especifica num_words como 10000 para limitar el vocabulario a las 10,000 palabras más frecuentes, 
# y oov_token se utiliza para representar palabras fuera del vocabulario.
```

```python
# fit_on_texts(): Este método ajusta el tokenizador a las textos de entrenamiento, 
# lo que significa que asigna un número entero único a cada palabra en el vocabulario.
tokenizer.fit_on_texts(reviews)
```

```python
# texts_to_sequences(): Convierte las frases de texto en secuencias de números enteros basadas en el vocabulario creado previamente.
sequences = tokenizer.texts_to_sequences(reviews)
```

```python
# pad_sequences(): Esta función ajusta las secuencias a una longitud fija de 100 (especificada en maxlen). 
# También rellena las secuencias con ceros al final (padding='post') 
# y trunca las secuencias más largas (truncating='post') si son necesarias.
padded_sequences = pad_sequences(sequences, maxlen=100, padding='post', truncating='post')
```

```python
# Creación del modelo

# Embedding Layer: La capa de embedding se utiliza para convertir las secuencias de números enteros en vectores densos (embeddings) de longitud 16. 
# La dimensión de entrada (input_dim) es 10000 (número de palabras en el vocabulario), 
# y la longitud de entrada (input_length) es 100 (longitud de secuencias).

model = keras.Sequential()
model.add(Embedding(input_dim=10000, output_dim=16, input_length=100))
```

```python
# Bidirectional LSTM Layer: Se utiliza una capa bidireccional LSTM con 64 unidades 
# para procesar las secuencias de texto. 
# La bidireccionalidad permite que la capa capture dependencias temporales en ambas direcciones.

model.add(Bidirectional(LSTM(64)))
```

```python
# Dense Layer: Una capa densa con una única unidad de salida y 
# función de activación sigmoide para realizar la clasificación binaria (positivo o negativo).
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compilación del modelo
# Compilación del Modelo: Se compila el modelo con el optimizador 'adam', 
# la función de pérdida 'binary_crossentropy' para problemas de clasificación binaria y se utiliza 'accuracy' como métrica de evaluación.
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Entrenamiento del modelo

# Entrenamiento del Modelo: El modelo se entrena con los datos de entrada (padded_sequences) 
# y las etiquetas (labels) durante 10 épocas, con una división del 20% para validación.
X_train, X_val, y_train, y_val = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)

X_train = np.array(X_train)
y_train = np.array(y_train)
X_val = np.array(X_val)
y_val = np.array(y_val)

# Entrena el modelo
model.fit(X_train, y_train, epochs=150, validation_data=(X_val, y_val))
```

```python
# Haciendo predicciones

test_reviews = ["Excelente película.", "No me gustó en absoluto."]
test_sequences = tokenizer.texts_to_sequences(test_reviews)
test_padded = pad_sequences(test_sequences, maxlen=100, padding='post', truncating='post')
predictions = model.predict(test_padded)
```

```python
for i, review in enumerate(test_reviews):
    print(f"Review: {review}")
    print(f"Sentimiento: {'Positivo' if predictions[i] > 0.5 else 'Negativo'}")
```

```python
# Tokenización y secuenciación de texto
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(reviews)
sequences = tokenizer.texts_to_sequences(reviews)
padded_sequences = pad_sequences(sequences, maxlen=100, padding='post', truncating='post')

# Crear el modelo
model = keras.Sequential()
model.add(Embedding(input_dim=10000, output_dim=16, input_length=100))
model.add(Bidirectional(LSTM(64)))
model.add(Dense(1, activation='sigmoid'))

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(padded_sequences, labels, epochs=10, validation_split=0.2)

# Realizar predicciones
test_reviews = ["Excelente película.", "No me gustó en absoluto."]
test_sequences = tokenizer.texts_to_sequences(test_reviews)
test_padded = pad_sequences(test_sequences, maxlen=100, padding='post', truncating='post')
predictions = model.predict(test_padded)

for i, review in enumerate(test_reviews):
    print(f"Review: {review}")
    print(f"Sentimiento: {'Positivo' if predictions[i] > 0.5 else 'Negativo'}")
```
